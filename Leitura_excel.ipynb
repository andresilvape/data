```python
import pandas as pd
```

```python
#leitura de arquivos excel
df1 = pd.read_excel("datasets_dio/Aracaju.xlsx")
df2 = pd.read_excel("datasets_dio/Fortaleza.xlsx")
df3 = pd.read_excel("datasets_dio/Recife.xlsx")
df4 = pd.read_excel("datasets_dio/Salvador.xlsx")
df5 = pd.read_excel("datasets_dio/Natal.xlsx")
```

```python
df["Data"]= df["Data"].astype("int64")
```

```python
# transformar para Date Time
df["Data"]= pd.to_datetime(df["Data"])
```

```python
df.dtypes
```

```python
#Agrupamento por Ano
df.groupby(df["Data"].dt.year)["Receita"].sum()
```

```python
df["Ano_Venda"] = df["Data"].dt.year
```

```python
df.sample(5)
```

```python
# Extrair o mes e o dia
df["mes_venda"], df["dia_venda"] = (df["Data"].dt.month, df["Data"].dt.day)
```

```python
df.sample(5)
```

```python
#data mais antiga
df["Data"].min()
```

```python
# diferenca de datas
df["diferenca_dias"] = df["Data"] - df["Data"].min()
```

```python
df.sample(5)
```

```python
#Criando coluna trimestre
df["Trimeste_Venda"] = df["Data"].dt.quarter
```

```python
df.sample(5)
```

```python
# Filtrando vendas de ano de um mes espec√≠fico
vendas_marco_2019 = df.loc[(df["Data"].dt.year == 2019) & (df["Data"].dt.month == 3)]
```
